```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import make_regression,load_iris
```

```python
load_iris().keys()
```

```python
dt=load_iris()
```

```python
dt.keys()
```

```python
df=pd.DataFrame(dt.data,columns=dt.feature_names)
```

```python
df.rename(columns={'sepal length (cm)':'sepal_length', 'sepal width (cm)':'sepal_width', 'petal length (cm)':'petal_length', 'petal width (cm)':'petal_width'}, inplace=True)
```

```python
df[df['sepal_length']>5]
```

```python
df
```

```python
df['target'] = dt.target
```

```python
df
```

```python
df.loc[df['sepal_length']<6,'new_column']='small sepals'
```

```python
df.loc[df['sepal_length']>=6,'new_column']='big sepals'
```

```python
df['updated_column']=df['sepal_length'].apply(lambda x: 'small sepals' if x<6 else 'large sepals')
```
